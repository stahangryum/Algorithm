## visualization

```python
def draw_two_circle(x1, y1, r1, x2, y2, r2):
    c1 = plt.Circle( (x1,y1), r1, fc = 'None', ec = 'b')
    c2 = plt.Circle( (x2,y2), r2, fc = 'None', ec = 'b')

    a = plt.axes(xlim=(-50, 50), ylim=(-50,50))
    a.add_patch(c1)
    a.add_patch(c2)

    a.set_aspect('equal')
    plt.show
```

```python
draw_two_circle(0, 0, 13, 40, 0, 37) # 두 점에서 만님
```

```python
draw_two_circle(0, 0, 3, 0, 7, 4) # 한 점에서 만남(접함)
```

```python
draw_two_circle(1, 1, 1, 1, 1, 5) # 만나지 않음
```

```python
draw_two_circle(3, 3, 30, 3, 3, 30) # 일치함
```

```python
from math import *

def num_intersections(x1, y1, r1, x2, y2, r2):
    
    dist_c = sqrt((x1-x2)**2 + (y1-y2)**2)
    sum_r = r1+r2
    diff_r = abs(r1-r2)

    if (x1, y1, r1) == (x2, y2, r2): return -1 # 일치
    elif (diff_r < dist_c < sum_r ): return 2 # 두 점에서 만난다.
    elif (sum_r == dist_c) or (diff_r == dist_c): return 1 # 한 점에서 만난다.
    elif (sum_r < dist_c) or (diff_r > dist_c): return 0 # 만나지 않는다.
    else: return('idk')

T = int(input())

C_list = []

for i in range(T):
    C_list.append(tuple(map(int, input().split())))

for C in C_list:
    print(num_intersections(*C))
```
